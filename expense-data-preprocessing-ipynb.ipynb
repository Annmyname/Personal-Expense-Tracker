# 2. Cleaning & Preprocessing data

## 2.1. General analysis

```python
# Check general information about dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.describe()
```

## 2.2. Variables 'Debit' & 'Credit'

```python
# Check if when Debit is not null, Credit must be null, and vice versa
check_rule = (
    ((df['Debit'].notna()) & (df['Credit'].isna())) | 
    ((df['Credit'].notna()) & (df['Debit'].isna()))
).all()

print("Does each customer use only one method (Debit or Credit)?", check_rule)
```

## 2.3. Variables 'Month Number'

```python
# Sort the dataframe by the 'Month Number' column in descending order 
df.sort_values(by = 'Month Number', ascending = False)
```

```python
# Ensure Date column is datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract month from Date
df['Computed Month'] = df['Date'].dt.month

# Check if Month Number matches Computed Month
check_month = (df['Month Number'] == df['Computed Month']).all()

print("Does 'Month Number' correctly represent the month from 'Date'?", check_month)
```

```python
# Drop variable 'Computed Month'
df = df.drop(columns=['Computed Month'])
```

# 2.4. Variable 'Sub-category' & 'Category'

```python
# Display all unique values in the 'Sub-category' column
print(df['Sub-category'].unique())
```

```python
# Group the dataframe by 'Category', 'Category Type', and 'Month Number'
# Then calculate the sum of 'Amount' for each group per month
cate_df = (
    df.groupby(['Category', 'Category Type', 'Month Number'])['Amount']
      .sum()
      .reset_index()
)

cate_df
```

## 2.5. Create variable 'Weekday' & 'Month Name'

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Extract the weekday name (e.g., Monday, Tuesday, ...)
df['Weekday'] = df['Date'].dt.day_name()

# Create a new column 'Month Name' with abbreviated month names (Jan, Feb, ...)
df['Month Name'] = df['Date'].dt.strftime('%b')
```
